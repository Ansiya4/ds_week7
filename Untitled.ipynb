```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data/titanic.csv')
```

```python
y= data['Survived']
x= data.drop(['Survived'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=101,stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier(random_state=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True,max_depth=4)
```

```python
train_accuracy=[]
test_accuracy=[]
for depth in range(1,10):
    model=DecisionTreeClassifier(max_depth=depth,random_state=10)
    model.fit(X_train,y_train)
    train_accuracy.append(model.score(X_train,y_train))
    test_accuracy.append(model.score(X_test,y_test))
```

```python
frame=pd.DataFrame({'max_depth':range(1,10),
                    'train_acc':train_accuracy,
                    'test_acc':test_accuracy,
                   })
frame.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['max_depth'],frame['train_acc'], marker='o',label="train_accuracy")
plt.plot(frame['max_depth'],frame['test_acc'], marker='o',label="test_accuracy")
plt.xlabel("Depth of tree")
plt.ylabel("Performance")
plt.legend()
```

```python
model=DecisionTreeClassifier(max_depth=8,max_leaf_nodes=25,random_state=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
